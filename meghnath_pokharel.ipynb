```python
# Defining the DNAMatching function
def DNAMatching(DNAStrand):
    """
    Function that finds the complementary DNA sequence of the corresponding DNA sequeces.
    
    Keyword arguments:
    A, T, C, G  -- base pairs
    
    Author: Meghnath Pokharel
    Date: 25-Feb-2020    
    """
    Keyword arguments:
    a,b -- sides of the right-angle triangle
    newdna = []                                  # New complementary DNA Strand

    for i in DNAStrand:                          # For loops   

        if i == 'G':                             # If there is 'G' in the coresponding DNA sequence, then it keeps adding the'C' which is the complementary base pair of  
            newdna.append('C')
            continue            
        if i == 'A':                             # If there is 'A' in the coresponding DNA sequence, then it keeps adding the'T' which is the complementary base pair of 'A'
            newdna.append('T')
            continue         
        if i == 'T':                              # If there is 'T' in the coresponding DNA sequence, then it keeps adding the'A' which is the complementary base pair of 'T'    
            newdna.append('A')
            continue       
        if i == 'C':                              # If there is 'C' in the coresponding DNA sequence, then it keeps adding the'G' which is the complementary base pair of 'C'     
            newdna.append('G')
            continue           
        else:                                      # If the function find the unknow base other than A, T, C, G, in the coresponding DNA sequence, then it breaks  
            print('Uknown base found:', i)
            break
    newdna = ''.join(newdna)
    return newdna
```

```python
#   
MatchingDNA = DNAMatching(dna)
```
