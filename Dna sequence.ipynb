```python

def DNAMatching(DNAStrand):

    newdna = []
    for i in DNAStrand:
        if i == 'G': 
            newdna.append('A')
            continue
        if i == 'A':
            newdna.append('G')
            continue
        if i == 'T':
            newdna.append('C')
            continue
        if i == 'C':
            newdna.append('A')
            continue
        else:
            print('Uknown base found:', i)
            break
    newdna = ''.join(newdna)
    return newdna
```

```python
MatchingDNA = DNAMatching(dna)
```
